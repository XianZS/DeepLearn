## 读取数据集

为了能用深度学习来解决现实世界的问题，我们经常从预处理原始数据开始， 而不是从那些准备好的张量格式数据开始。 在Python中常用的数据分析工具中，我们通常使用pandas软件包。 像庞大的Python生态系统中的许多其他扩展包一样，pandas可以与张量兼容。


```python
import pandas as pd
import torch.distributed.rpc

df=pd.read_csv("./data_people.csv")
print(df.head())
```

## 处理缺失值

用pandas处理缺失的数据时，我们可根据情况选择用插值法和删除法。

```python
# 切分数据集
inputs,outputs=df.iloc[:,1:2],df.iloc[:,2:]
outputs
```

```python
# 替换outputs之中的缺失数值
# (1)[平均值替换]替换年龄缺失值
mean_age=int(outputs["age"].mean())
print(mean_age,type(mean_age))
outputs["age"]=outputs["age"].fillna(mean_age)
# (2)[默认数值替换]替换性别缺失值
outputs["city"]=outputs["city"].fillna("上海")
outputs
```

## 替换为张量模式

```python
res=outputs.iloc[:,2]
torch.tensor(res.to_numpy(dtype=float))
```
